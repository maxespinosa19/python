# Simple Linear Regression - Housing Prices
We will try to predict the price of a house based on the area in sq. ft.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

# Import the Data

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data.head()
```

```python
data.describe()
```

## Let's begin the regression
Declare dependent and independent variables 

```python
y = data['price']
x = data['size']
```

## Let's explore the data

```python
plt.scatter(x, y)
plt.xlabel('Size', fontsize=20)
plt.ylabel('Price', fontsize=20)
```

## Regression itself

```python
x1 = sm.add_constant(x)
results = sm.OLS(y, x1).fit()
results.summary()
```

```python
plt.scatter(x,y,c='green')
plt.xlabel('Size', fontsize=20)
plt.ylabel('Price', fontsize=20)
y_predicted= x * 223.1782 + 101900
fig = plt.plot(x, y_predicted, lw= 3, c='pink')
```
